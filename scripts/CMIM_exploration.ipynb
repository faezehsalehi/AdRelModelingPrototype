```python
import CMIM
import sys
import pandas as pd
import numpy as np
import sklearn
from sklearn.feature_extraction import DictVectorizer
```

```python
n_cols = ["impressions", "ad_campaign", "category", "game_audience", 
          "gender", "country","region", "reachability", "device_type", 
          "bid_value", "ad_type", "media_type", "asset_size", "os", "has_install", "pub_app"]

df = pd.read_csv("data/000000_0",sep = '\t', header = None, names = n_cols)

df[["ad_campaign", "category", "game_audience", "gender", "country", "region", 
    "reachability","device_type", "ad_type", "media_type","asset_size", "os", 
    "has_install", "pub_app"]] = df[["ad_campaign", "category", "game_audience", "gender", 
                          "country", "region", "reachability","device_type", 
                          "ad_type", "media_type","asset_size", "os", "has_install", "pub_app"]].astype("str")
df[["impressions"]] = df[["impressions"]].astype("int")
df[["bid_value"]] = df[["bid_value"]].astype("float")
```

```python
selected_campaigns = []
for line in open("selected_campaigns_for_modeling.csv","r"):
    selected_campaigns.append(line.replace('"','').strip())
```

```python
df_selected = df[df['ad_campaign'].isin(selected_campaigns)]
```

```python
df2 = df_selected.loc[np.repeat(df_selected.index.values, df_selected.impressions)]
```

```python
df2 = df2.reset_index(drop=True).head(100000)
```

```python
def one_hot_dataframe_train(data, cols, replace=True):
    """ Takes a dataframe and a list of columns that need to be encoded.
    Returns the data and the fitted vectorizor.
    """
    vec = DictVectorizer(dtype=np.int8)
    vecData = vec.fit_transform(data[cols].to_dict(orient='records')).toarray()
    feature_names = vec.get_feature_names()

    return vecData, np.array(feature_names)
```

```python
xtrain_Arr, feature_names = one_hot_dataframe_train(df2, ["category", "game_audience", "gender", "country", "region", 
    "reachability","device_type", "ad_type", "media_type","asset_size", "os"])
```

```python
converted_df = pd.DataFrame(xtrain_Arr, columns = feature_names,dtype=np.float64)
y = y.astype("int")
```
